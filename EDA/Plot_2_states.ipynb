```python
# import geopandas as gpd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import json
import os
import pandas as pd
import seaborn as sns
```

```python
dir_ = "data/PA_data/"
df_pa = pd.read_csv("data/PA_data/plan_metrics_PA_PRES16.csv")
df_va = pd.read_csv("data/VA_data/plan_metrics_ATG17.csv")
df_pa.head()
```

```python
df_va.head()
```

```python
sns.distplot( df_pa["mms"] , color="#2ecc71")
sns.distplot( df_va["mms"] , color= "orange", axlabel="Mean-Medians across plans")
plt.legend(labels=['PA', 'VA'])
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
fig.savefig('mms.png', dpi=100)
```

```python
sns.distplot( df_pa["egs"] , color="#2ecc71")
sns.distplot( df_va["egs"] , color="orange", axlabel="Efficiency gaps across plans")
plt.legend(labels=['PA', 'VA'])
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
fig.savefig('egs.png', dpi=100)
```

```python
sns.distplot( df_pa["hmss"] , color="#2ecc71")
sns.distplot( df_va["hmss"] , color="orange", axlabel="Seats won by Democratic Party across plans")
plt.legend(labels=['PA', 'VA'])
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
fig.savefig('hmss.png', dpi=100)
```

```python
sns.distplot( df_pa["nb_cuts"] , color="#2ecc71")
sns.distplot( df_va["nb_cuts"] , color="orange", axlabel="Number of Cuts across plans")
plt.legend(labels=['PA', 'VA'])
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
fig.savefig('nb_cuts.png', dpi=100)
```
